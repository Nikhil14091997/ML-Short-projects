```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
```

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
df_cancer.tail()
```

```python
df_cancer.head()
```

```python
from sklearn.model_selection import train_test_split
X = df_cancer.drop(['target'], axis = 1)
y = df_cancer['target']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
y_test.shape
```

```python
X_train_min = X_train.min()
X_train_max = X_train.max()
X_train_range = (X_train_max - X_train_min)
X_train_scaled = (X_train - X_train_min)/X_train_range
X_train.head()
```

```python
X_train_scaled.head()
```

```python
X_test_min = X_test.min()
X_test_max = X_test.max()
X_test_range = (X_test_max - X_test_min)
X_test_scaled = (X_test - X_test_min)/X_test_range
print(X_test.head())
```

```python
print(X_test_scaled.head())
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
svc_model_1 = SVC()
svc_model_1.fit(X_train, y_train)
y_pred_1 = svc_model_1.predict(X_test)
cm_1 = confusion_matrix(y_test, y_pred_1)
print(np.array(cm_1))
accuracy_score(y_test, y_pred_1)
#print(classification_report((y_test, y_pred_1))
```

```python
svc_model_2 = SVC()
svc_model_2.fit(X_train_scaled, y_train)
y_pred_2 = svc_model_2.predict(X_test_scaled)
cm_2 = confusion_matrix(y_test, y_pred_2)
print(np.array(cm_2))
accuracy_score(y_test, y_pred_2)
```

```python
param_grid= {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
# fitting the model for grid search
grid.fit(X_train_scaled, y_train)
```

```python
# print best parameter after tuning
print("Printing Best Parameters: \n", grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python

grid_predictions = grid.predict(X_test_scaled)
cm_3 = confusion_matrix(y_test, grid_predictions)
print(np.array(cm_3))
accuracy_score(y_test, grid_predictions)
```
